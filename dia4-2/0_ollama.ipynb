# Ejercicio

Haz un programa que reciba como entrada una petición de un usuario en lenguaje natural de generación de código en Python (ejemplo, "dame un programa que ordene un array") y devuelva un programa en Python que cumpla esa función. Para ello, haz lo siguiente:

````python
import requests
import re
import traceback


def crear_prompt(peticion_usuario, error_anterior=None):

    prompt = f"""
Eres un asistente que genera código Python ejecutable.

Genera un programa en Python que haga lo siguiente:
{peticion_usuario}

Devuelve únicamente el código dentro de un bloque ```python```.
"""

    if error_anterior:
        prompt += f"""

El código anterior produjo el siguiente error:

{error_anterior}

Corrige el programa para que funcione.
Devuelve solo código Python.
"""

    return prompt


def llamar_llm(prompt):

    url = "http://localhost:11434/api/generate"

    payload = {
        "model": "llama3",
        "prompt": prompt,
        "stream": False
    }

    response = requests.post(url, json=payload)

    data = response.json()

    return data["response"]


def extraer_codigo(respuesta_llm):

    patron = r"```python(.*?)```"

    match = re.search(patron, respuesta_llm, re.DOTALL)

    if match:
        return match.group(1).strip()

    return None


def ejecutar_codigo(codigo):

    try:
        exec(codigo)
        return True, None

    except Exception:
        error_completo = traceback.format_exc()
        return False, error_completo


def generar_y_ejecutar_codigo(peticion_usuario):

    error_anterior = None
    max_iteraciones = 5

    for i in range(max_iteraciones):

        print(f"\nIteración {i+1}")

        prompt = crear_prompt(peticion_usuario, error_anterior)

        respuesta_llm = llamar_llm(prompt)

        codigo = extraer_codigo(respuesta_llm)

        if not codigo:
            print("No se encontró código en la respuesta.")
            return

        print("\nCódigo generado:\n")
        print(codigo)

        exito, error = ejecutar_codigo(codigo)

        if exito:
            print("\nEl código se ejecutó correctamente.")
            return

        else:
            print("\nError detectado:")
            print(error)
            error_anterior = error

    print("\nNo se pudo generar un código funcional tras 5 intentos.")


def main():

    peticion_usuario = input("Introduce la petición de código: ")

    generar_y_ejecutar_codigo(peticion_usuario)


if __name__ == "__main__":
    main()
````
